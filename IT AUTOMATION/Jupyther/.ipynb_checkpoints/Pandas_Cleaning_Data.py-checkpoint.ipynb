```python
import pandas as pd
import numpy as np
```

```python
people ={
    'first':['Carey','Jane','john','Chris',np.nan,None,'NA'],
    'last':['Shelby','Doe','Doe','Schafer',np.nan,np.nan,'Missing'],
    'email':['carey@xxx.com','jane@xxx.com','john@xxx.com',None,np.nan,'anonymous@xxx.com','NA'],
    'age':['33','55','63','36',None,None,'Missing']
}
```

```python
df = pd.DataFrame(people)
df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis='index',how='all')
```

```python
df.dropna(axis='columns',how='all')
```

```python
df.dropna(axis='index',how='any',subset=['email'])
```

```python
df.dropna(axis='index',how='all',subset=['email'])
```

```python
df.dropna(axis='index',how='all',subset=['last','email'])
```
