```python
import folium
```

```python
countries_geojson = 'https://raw.githubusercontent.com/johan/world.geo.json/master/countries.geo.json'
```

```python
country_survey = refine_data.Country.value_counts()
country_survey_df = pd.DataFrame({'Country':country_survey.index, 'Count':country_survey.values})
country_survey_df
```

```python
country_survey_df.at[0,"Country"] = "United States of America"
country_survey_df.at[12,"Country"] = "Russia"
```

```python
m = folium.Map(location=[30,0],zoom_start=2)
```

```python
folium.Choropleth(
    geo_data=countries_geojson,
    data=country_survey_df,
    columns=["Country", "Count"],
    key_on="feature.properties.name",
    fill_color="OrRd",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Respondents",
).add_to(m)

m
```
