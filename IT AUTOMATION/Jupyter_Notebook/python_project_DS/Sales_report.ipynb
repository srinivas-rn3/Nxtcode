```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {'Salesperson': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Quarter': [1, 2, 3, 4, 1],
    'Sales': [10000, 15000, 12000, 18000, 9000]}
```

```python
df = pd.DataFrame(data)
```

```python
print("Sales Data")
print(df)
```

```python
# Basic statistics
print("\nBasic Statistics:")
print(df.describe())
```

```python
# Total sales per quarter
print("Total sales per quater:")
quaterly_sales = df.groupby('Quarter')['Sales'].sum()
print("\nTotal Sales per Quarter:")
print(quaterly_sales)
```

```python
#Visualizing the data
quaterly_sales.plot(kind='bar',rot=0)
plt.title("Total sales Quarter")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Generate sample data
np.random.seed(0) # # For reproducibility
salespeople = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack']
years = np.random.choice(range(2015,2023),size=100)
quarters = np.random.choice([1,2,3,4],size=100)
sales = np.random.randint(5000,20000,size=100)
```

```python
data= {
    'Salesperson': np.random.choice(salespeople,size=100),
    'Year': years,
    'Quarters': quarters,
    'Sales' : sales
}
```

```python
df_large = pd.DataFrame(data)
#display the fisrt fwe rows of the large Dataframe
print("Large Sales Data:")
print(df_large)
```

```python
# Basic statistics
print("\nBasics  Statistian:")
print(df_large.describe())
```

```python
# Total sales per quarter
large_quarterly_sales = df_large.groupby("Quarters")["Sales"].sum()
print("\nTotal sales per Quaters")
print(large_quarterly_sales)
```

```python
# Visualizing the data
large_quarterly_sales.plot(kind='bar',rot=0)
plt.title("Large Sales Report")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.show()
```
