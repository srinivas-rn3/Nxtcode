***Exploratory Data Analysis (EDA)***

```python
# Create a bar plot to visualize commodity categories in data2
plt.figure(figsize=(10,6))
sns.countplot(data=data2,x='category',palette='viridis')
plt.xticks(rotation=45)
plt.title("Distribution of Commodity Categories in data2",fontsize= 14 ,
          fontweight='bold',color='magenta')
plt.xlabel("Category",fontsize=12,fontweight='bold',color='darkblue')
plt.ylabel("Count",fontsize=12,fontweight='bold',color='darkblue')
plt.savefig("Distribution of Commodity Categories in data2.png")
# Print the number of unique categories
unique_categories = data2['category'].nunique()
plt.text(0.7,0.9,f"Unique Categories:{unique_categories}",transform=plt.gca().transAxes)
plt.show()
```

The bar plot shows the distribution of commodity categories in data2. It's evident that there are multiple categories, and you can see the count of each category. This can help in understanding the composition of the data.

***Feature Engineering***

```python
# Calculate duration between 'start_date' and 'end_date'
data1['duration'] = (data1['end_date'] - data1['start_date']).dt.days
```

```python
# Create a histogram to visualize the duration
plt.figure(figsize=(10,6))
sns.histplot(data=data1,x='duration',bins=20,kde=True,color='green')
plt.title("Distribution of Duration between Start and End Dates in data1.png",
         fontsize=12,fontweight='bold',color='darkblue')
plt.xlabel("Duration (days)",fontsize = 12,fontweight = 'bold',color = 'darkblue')
plt.ylabel("Frequency",fontsize = 12,fontweight='bold',color = 'darkblue')
plt.savefig("Distribution of Duration between  Start and  End Dates in date1.png")

# Print summary statistics
mean_duration = data1['duration'].mean()
mediun_duration = data1['duration'].median()
plt.text(0.7, 0.9, f"Mean Duration: {mean_duration:.2f} days",transform = plt.gca().transAxes)
plt.text(0.7, 0.8, f"Median Duration:{mediun_duration:.2f} days",transform = plt.gca().transAxes)


plt.show()
```

The histogram shows the distribution of the duration between 'start_date' and 'end_date' in data1. It provides insights into the range and central tendency of this duration.

***Correlation Analysis***

```python
# Select only the numeric columns for correlation analysis
numeric_data1 = data1.select_dtypes(include='number')
# Create a heatmap to visualize correlations
# Calculate correlations between numeric columns in data1
correlation_data1 = numeric_data1.corr()
```
