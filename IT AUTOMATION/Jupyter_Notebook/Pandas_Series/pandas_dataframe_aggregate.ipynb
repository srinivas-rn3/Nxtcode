***Pandas dataframe.aggregate()***

(https://www.geeksforgeeks.org/python-pandas-dataframe-aggregate/?ref=lbp)

Dataframe.aggregate() function is used to apply some aggregation across one or more column. Aggregate using callable, string, dict, or list of string/callables. Most frequently used aggregations are:<br>

sum: Return the sum of the values for the requested axis<br>
min: Return the minimum of the values for the requested axis<br>
max: Return the maximum of the values for the requested axis<br>

Syntax: DataFrame.aggregate(func, axis=0, *args, **kwargs)<br>

Parameters:<br>
func : callable, string, dictionary, or list of string/callables. Function to use for aggregating the data. If a function, must either work when passed a DataFrame or when passed to DataFrame.apply. For a DataFrame, can pass a dict, if the keys are DataFrame column names.<br>

axis : (default 0) {0 or ‘index’, 1 or ‘columns’} 0 or ‘index’: apply function to each column. 1 or ‘columns’: apply function to each row.<br>

Returns: Aggregated DataFrame

```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv(r"C:/Users/srini/Downloads/nba.csv")
df = pd.DataFrame(data)
df[:10]
```

```python
#Aggregation works with only numeric type columns.
# Applying aggregation across all the columns  
# sum and min will be found for each  
# numeric type column in df datafram
df.aggregate({'Number':['sum','min'],
            'Age':['max','min'],
             'Salary':['sum']})
```

```python
df.aggregate({"Weight":['min','max']})
```
